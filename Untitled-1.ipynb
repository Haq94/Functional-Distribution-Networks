```python
import numpy as np

def sample_function():
    fn_type = np.random.choice(["sine", "quadratic", "step"])
    if fn_type == "sine":
        a = np.random.uniform(0.5, 2.0)
        b = np.random.uniform(1.0, 3.0)
        return lambda x: a * np.sin(b * x)
    elif fn_type == "quadratic":
        a = np.random.uniform(-1.0, 1.0)
        b = np.random.uniform(-1.0, 1.0)
        return lambda x: a * x**2 + b
    else:  # step function
        return lambda x: np.where(x > 0, 1.0, -1.0)

def generate_dataset(n_tasks=1000, n_points=20, x_range=(-3, 3)):
    X, Y = [], []
    for _ in range(n_tasks):
        f = sample_function()
        x = np.random.uniform(*x_range, size=(n_points, 1))
        y = f(x)
        X.append(x)
        Y.append(y)
    return X, Y
```

```python
X,Y = generate_dataset()

import matplotlib.pyplot as plt

plt.plot(X(:,0,0),Y(:,0,0))
```
